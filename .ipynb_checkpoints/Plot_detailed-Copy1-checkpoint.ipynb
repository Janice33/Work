```python
name="Hidden Brain"
name=name.replace(' ','-').lower()
print(name)
```

```python
podcast="https://chartable.com/podcasts/"+name+"/charts"
driver.get(podcast)
```

```python
final=[]
table = driver.find_element_by_xpath('//*[contains(@class,"w-100 f5 mb1")]//tr')
for item in table.find_elements_by_xpath('//td//a[@class="link blue"]'):
    final.append(item.get_attribute('href'))
for f in final:
    print(f)
```

```python
spotify=[]
itunes=[]
for f in final:
    if "spotify" in f:
        spotify.append(f)
    elif "itunes" in f:
        itunes.append(f)
        
print(len(spotify))
print(len(itunes))
```

```python
import re
spotify_final=[]
for link in spotify:
    pattern="spotify\/(.*?)$"
    crop = re.search(pattern, link)
    if crop:
        spotify_final.append(crop.group(1))
print(spotify_final)
```

```python
import re
itunes_final=[]
for link in itunes:
    pattern="itunes\/(.*?)-podcasts-?$"
    crop = re.search(pattern, link)
    if crop:
        itunes_final.append(crop.group(1))
print(itunes_final)
```

```python
#Apple
itunes_plot = []

for key in itunes_final:
    url = 'https://chartable.com/api/podcasts/'+name+'/rank_history?chart_id='+key+'-'+'podcasts&chart_type=itunes'
    driver.get(url)
    plot = driver.find_element_by_tag_name('pre')
    itunes_plot.append([key,plot.text])
```

```python
import pandas as pd
df_itunes = pd.DataFrame(itunes_plot, columns =['Country-Genre', 'Plot_points'])  
df_itunes.head(20)
```

```python
#Spotify Podcasts
spotify_plot = []

for key in spotify_final:
    url = 'https://chartable.com/api/podcasts/'+name+'/rank_history?chart_id='+key+'&chart_type=spotify'
    driver.get(url)
    plot = driver.find_element_by_tag_name('pre')
    spotify_plot.append([key,plot.text])
```

```python
import pandas as pd
df_spotify = pd.DataFrame(spotify_plot, columns =['Country-Genre', 'Plot_points'])  
df_spotify.head(20)
```
